```python
import pandas as pd
import seaborn as sns

tmbd = pd.read_csv("./Base de Dados/tmdb_5000_movies.csv")
tmbd.head()
```

```python
linguasQuantidade = tmbd.original_language.value_counts().to_frame().reset_index()
linguasQuantidade.columns = ["Lingua_Original", "Quantidade"]
linguasQuantidade.head()
```

```python
sns.catplot(tmbd, x = "original_language", kind="count")
```
